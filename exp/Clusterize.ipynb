```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import copy
import math
from collections import defaultdict

class dotdict(dict):
    __getattr__ = dict.get
    __setattr__ = dict.__setitem__
    __delattr__ = dict.__delitem__
```

```python
fig = plt.figure()
ax = plt.axes()
```

```python
def load_points(file):
    f = open(file)
    next(f)
    points = []
    for point in f.read().splitlines():
        points.append({'x': int(point.split(' ')[0]), 'y': int(point.split(' ')[1])})
    f.close();
    return points
```

```python
def add_clusters_to_points(cluster_file, points):
    file = open(cluster_file)
    for i, cluster in enumerate(file.read().splitlines()):
        points[i]['cluster'] = cluster
```

```python
points = load_points('dataset-1')
add_clusters_to_points('kruskal_dataset-1', points)
```

```python
def plot_points(points):
    p_x = []
    p_y = []

    point_by_clusters = defaultdict(list)
    for point in points: 
        point_by_clusters[point['cluster']].append(point)

    clusters = defaultdict(lambda: defaultdict(list))

    for c, point in point_by_clusters.items():
        for p in point:
            clusters[c]['x'].append(p['x'])
            clusters[c]['y'].append(p['y'])

    for cluster in clusters:
        plt.plot(clusters[cluster]['x'], clusters[cluster]['y'], marker='.', linestyle='')
```

```python
plot_points(points)
```
